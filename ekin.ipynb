```python
def gendata(loc=0):
    population= us_data         #United States
    sample = hou_data             #Houston
    
    
    population = stats.norm.rvs(size=1000, random_state=42)
    sample = stats.norm.rvs(loc=loc, size=200, random_state=42)
    
    
  
  
    plt.subplot(2, 1, 1)
    plt.scatter(range(len(population)), population, label="United States")
    plt.scatter(range(len(sample)), sample, label="Houston,TX")
    plt.legend()
    
    
    
    plt.subplot(2, 1, 2)
    plt.hist(population, 20, density=True, alpha=0.7, label="United State")
    plt.hist(sample, 20, density=True, alpha=0.7, label="Houston,Tx")
    plt.axvline(population.mean(), color='k', linestyle='dashed', linewidth=1)
    plt.axvline(sample.mean(), color='k', linestyle='dashed', linewidth=1)
    plt.legend()  
    
    return population , sample
```

```python
population, sample = gendata()
```

```python
 stats.ttest_1samp(sample, population.mean())
```
